### #1.2

Crea una funcin similar a la que se pide en el ejercicio 6 de la segunda tanda de ejercicios del workout. Es decir, una funcin para pintar diagramas de dispersion que tenga seis argumentos, pero ojo al cambio en el quinto:
1. Un dataframe
2. Una columna numrica del dataframe
3. Una columna numrica del dataframe diferente a la anterior.
4. Un columna categrica (Opcional, valor por defecto None)
5. Un valor para el tamao de los puntos o una cadena, pero que puede ser una cadena de caracteres, en cuyo caso se debe interpretar como una de las columnas del dataframe cuyos valores deben controlar el tamao
6. Un sexto argumento "escala", que permita manipular la escala del tamao en el caso de que el quinto argumento sea una cadena. Este argumento multiplicar a los valores de la columna indicada en el parmetro anterior (as si quiero dividir por 10000, este sexto argumento tendr que valer 1/10000)

La funcin debe pintar el diagrama de dispersin de la columna numrica 1, en el eje X, frente a la columna numrica 2, en el eje Y, usando un cdigo de colores en funcin de la columna categrica y con el tamao de los puntos segn el argumento correspondiente. 

Si no se diera una columna categrica como argumento, el color no debe pasarse como parmetro a la funcin de Seaborn o Matplotlib que uses.

EXTRA: Aade un argumento que controle si la funcin pinta una leyenda para colores y tamaos o un colorbar y una leyenda para los tamaos. NO TE PREOCUPES POR EL FORMATO DE LA LEYENDA (lo veremos en la sesin en vivo)

```python
def scatter_custom(
    df: pd.DataFrame,
    x: str,
    y: str,
    hue: str | None = None,
    size: int | float | str = 40,
    escala: float = 1.0,
    legend_mode: str = "legend",  # "legend" | "colorbar"
    cmap: str = "viridis",
    title: str | None = None,
):
    # --- validacin de columnas ---
    campos = [x, y]
    if isinstance(size, str):
        campos.append(size)
    if hue is not None:
        campos.append(hue)
    faltan = [c for c in campos if c not in df.columns]
    if faltan:
        raise KeyError(f"Las columnas no existen en el DataFrame: {faltan}")

    d = df.copy()

    # tamao: fijo vs columna
    if isinstance(size, str):
        d["_size_"] = pd.to_numeric(d[size], errors="coerce").fillna(0) * escala
        size_col, size_fixed = "_size_", None
    else:
        size_col, size_fixed = None, float(size)

    fig, ax = plt.subplots()

    if hue is None:
        ax.scatter(d[x], d[y], s=(d[size_col] if size_col else size_fixed), alpha=0.85)
        ax.set_xlabel(x); ax.set_ylabel(y)
        if title: ax.set_title(title)
        plt.show()
        return ax

    hue_is_numeric = pd.api.types.is_numeric_dtype(d[hue])

    if legend_mode == "colorbar" and hue_is_numeric:
        sc = ax.scatter(
            d[x], d[y],
            c=d[hue], cmap=cmap,
            s=(d[size_col] if size_col else size_fixed),
            alpha=0.85
        )
        cbar = plt.colorbar(sc, ax=ax); cbar.set_label(hue)

        if size_col:
            vals = d[size_col].replace(0, np.nan).dropna()
            if not vals.empty:
                for q in np.quantile(vals, [0.25, 0.5, 0.75]):
                    ax.scatter([], [], s=q, c="gray", alpha=0.6, label=f"{q:.2g}")
                ax.legend(title="Tamao", frameon=True, loc="best")
    else:
        if size_col:
            lo = d[size_col].replace(0, np.nan).min()
            hi = d[size_col].max()
            lo = 10 if not np.isfinite(lo) else max(10, lo)
            hi = 200 if not np.isfinite(hi) or hi <= lo else hi
            sns.scatterplot(data=d, x=x, y=y, hue=hue, size=size_col, sizes=(lo, hi), alpha=0.85, ax=ax)
        else:
            sns.scatterplot(data=d, x=x, y=y, hue=hue, s=size_fixed, alpha=0.85, ax=ax)

        if legend_mode == "colorbar" and hue_is_numeric:
            print("Nota: hue es numrico; usa legend_mode='colorbar' para colorbar.")

    ax.set_xlabel(x); ax.set_ylabel(y)
    if title: ax.set_title(title)
    plt.show()
    return ax
```

Previamente crea una columna nueva "log_poblacion" y asignale el log decimal de la poblacion (np.log10() y como argumento los valores de la columna "population_total"). Usa "log_poblacion" para el color.

Un ejemplo de como podra salir para que lo tengas de referencia: (olvidate de la leyenda, auque si quieres como extra puedes intentar ver cmo crearla as)

```python
# 1) Columna logartmica de poblacin (evita log(0))
df = df.copy()
df["log_poblacion"] = np.log10(df["population_total"].replace(0, np.nan))

# 2) Elegimos el tamao de la ciudad para el tamao del punto
#    (usa rea total km; si prefieres, cambia a 'area_land_km2')
size_col = "area_total_km2"
tam = df[size_col].fillna(0) * 0.8  # escala para que se vean bien

# 3) Dispersin: X=longitud, Y=latitud, color=log_poblacion, tamao=rea
fig, ax = plt.subplots()
sc = ax.scatter(
    df["longd"],           # eje X: longitud
    df["latd"],            # eje Y: latitud
    s=tam,                 # tamao por rea
    c=df["log_poblacion"], # color por log10 de poblacin
    alpha=0.85
)

# 4) Colorbar y etiquetas
cbar = plt.colorbar(sc, ax=ax)
cbar.set_label("log10(poblacin)")

ax.set_xlabel("longitud")
ax.set_ylabel("latitud")
ax.set_title("Ciudades: longitud vs latitud (tamao=rea, color=log10 poblacin)")

plt.show()
```

```python
plt.rcParams['figure.dpi'] = 120

def save_fig(path):
    plt.savefig(path, bbox_inches='tight', dpi=180)
    plt.close()

def load_titanic():
    """Carga el dataset Titanic de seaborn; si no est disponible, genera uno sinttico til para las figuras."""
    try:
        df = sns.load_dataset('titanic').copy()
        need = {'survived','sex','who','class','embark_town','fare'}
        if need.issubset(df.columns):
            return df
    except Exception:
        pass
    # Fallback sinttico
    rng = np.random.default_rng(42)
    n = 800
    sexes = rng.choice(['male','female'], size=n, p=[0.65, 0.35])
    who = np.array(['man' if s=='male' else 'woman' for s in sexes])
    child = rng.random(n) < 0.10
    who[child] = 'child'
    classes = rng.choice(['First','Second','Third'], size=n, p=[0.25,0.20,0.55])
    ports = rng.choice(['Cherbourg','Queenstown','Southampton'], size=n, p=[0.2,0.1,0.7])
    base = np.where(classes=='First', 0.62, np.where(classes=='Second', 0.47, 0.26))
    boost = np.where(who=='woman', 0.22, np.where(who=='child', 0.18, 0.0))
    surv_p = np.clip(base + boost, 0, 0.98)
    survived = (rng.random(n) < surv_p).astype(int)
    fare = (np.where(classes=='First', rng.normal(85, 30, n),
             np.where(classes=='Second', rng.normal(30, 10, n), rng.normal(12, 6, n)))).clip(0, None)
    return pd.DataFrame({
        'survived': survived,
        'sex': sexes,
        'who': who,
        'class': classes,
        'embark_town': ports,
        'fare': fare,
    })

df = load_titanic()
df.head()
```
